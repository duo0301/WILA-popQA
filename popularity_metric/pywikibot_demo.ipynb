# Pywikibot Demo

```python
import pywikibot
```

```python
site = pywikibot.Site("wikidata", "wikidata")
repo = site.data_repository()
    
# Fetch an item
# About ItemPage class: https://doc.wikimedia.org/pywikibot/master/api_ref/pywikibot.html#pywikibot.ItemPage
q_number = "Q12866753"
item = pywikibot.ItemPage(repo, q_number)
```

## Popularity metrics

- Sitelinks: inter-wiki links, number of different language versions, “measure of canonicity”
- Linked pages: intra-wiki links, pages the page links to, “outdegree”
- Backlinks: intra-wiki links, pages that link to the page, “indegree”


```python
# Sitelinks: https://doc.wikimedia.org/pywikibot/master/api_ref/pywikibot.page.html#page.ItemPage.iterlinks
# Linked pages: https://doc.wikimedia.org/pywikibot/master/api_ref/pywikibot.page.html#page.BasePage.linkedPages
# Backlinks: https://doc.wikimedia.org/pywikibot/master/api_ref/pywikibot.page.html#page.BasePage.backlinks

for sitelink_counter, sitelink in enumerate(item.iterlinks("wikipedia"), start=1): 
    print(f"Retrieving sitelink {sitelink_counter} for {item}")
    print(sitelink)
    print(f"Retrieving linked pages for {sitelink}")
    for linked_page in sitelink.linkedPages():
        print(linked_page)
    print(f"Retrieving backlinks for {sitelink}")
    for backlink in sitelink.backlinks():
        print(backlink)
else: 
    print(f"Number of sitelinks: {sitelink_counter}")
    
```

```python
# Alternative way to get sitelinks (not restricted to Wikipedia) 
# item.iterlinks() or via attribute sitelinks
for sitelink in item.sitelinks:
    print(sitelink)
```

## Properties

```python
# About claims attribute: https://doc.wikimedia.org/pywikibot/master/api_ref/pywikibot.page.html#page.ItemPage.DATA_ATTRIBUTES
# Print all properties associated with an item
for claim in item.claims:
    print(claim)
```

```python
# Get statements involving P19
item.claims["P19"]
```

```python
# Get claim for a specific statement
# About getTarget() method: https://doc.wikimedia.org/pywikibot/master/api_ref/pywikibot.page.html#page.Claim.getTarget

pob_item = item.claims["P19"][0].getTarget() # getTarget returns ItemPage object
pob_item.labels.get('en', "") # get english label of ItemPage object
```
